Q.5.Show only the Titles of all TV Shows that were released in India only.

```python
import matplotlib.pyplot as plt
```

```python
filtered_data = data[(data['Category'] == 'TV Show') & (data['Country'] == 'India')]
```

```python
# Count the occurrences of each title
title_counts = filtered_data['Title'].value_counts()
```

```python
# Plot the bar graph
plt.figure(figsize=(12, 8))
title_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Title')
plt.ylabel('Count')
plt.title('Count of TV Shows in India')
plt.xticks(rotation=90)
plt.show()
```
